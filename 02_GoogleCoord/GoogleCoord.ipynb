```python
#get all the imports

import pandas as pd
import numpy as np

#we will use selenium webdriver
from selenium import webdriver
from time import sleep
```

```python
#create webdriver instance
driver = webdriver.Chrome()

#create an empty dataframe
df = pd.DataFrame()

#create empty lists 
latitude= []
longitude = []
at_this_location =[]

#get the url
driver.get("https://www.google.com/maps")
sleep(2)

#list of all addresses

addresses = ['Yorkshire Bank Plc, 1a Peel Square, S70 2PL',
'50 Lisson Street, NW1 5DF',
'THIRD FLOOR, 50 Lisson Street, NW1 5DF',
'UNIT 1, 50 Lisson Street, NW1 5DF',
'UNIT 3, 50 Lisson Street, NW1 5DF',
'UNIT 4 - 3RD FLOOR, 50 Lisson Street, NW1 5DF',
'22 Saville Row, NE1 8JE',
'Unit 16, First Avenue, DH2 1AG',
'Morse Point Menswear Ltd, Royal Square, LA23 3ED',
'Punch Bowl, Old Bath Road, SL3 0PH']


#loop through each address to find its coordinates.
for i,addr in enumerate(addresses):
    
    #firstly initialize the location list with NaNs
    at_this_location.append(np.nan)
    
    
    Place = driver.find_element("id", "searchboxinput")
    Place.send_keys(addr)
    sleep(2)
    Submit = driver.find_element("id",
        "searchbox-searchbutton")
    Submit.click()
    sleep(2)
    url = driver.current_url
    print(url)
    
    #get the index at which we find "@" symbol
    index = url.index("@")
    
    #create a substring starting from "@" symbol to next 21 values
    coord = url[index+1:index+1+21]
    
    print(coord)
    
    #get the index at which we find ',' symbol
    index_comma = coord.index(",")
    
    #separate the first and second substring by ',' symbol
    first = coord[0:index_comma].replace(',','')
    second = coord[index_comma+1:].replace(',','')
    
    latitude.append(first)
    longitude.append(second)
    
    #now check for at this location
    
    try:
        
        ele = driver.find_element("xpath","/html/body/div[3]/div[9]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div")
        
        if ele.get_attribute("class") =='m6QErb WNBkOb':
            
            for j in range(0,20):
                try:
                    ele = driver.find_element("xpath",f'/html/body/div[3]/div[9]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[{j}]')
                    
                    if ele.text == "At this location" or ele.text == "At this place":
        
                        location = ""  #creating an empty string to collect the places without any empty space " " in between
                        
                        for inside_div in range(1,5):
                            try:
                                ele = driver.find_element("xpath",f'/html/body/div[3]/div[9]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[{j+1}]/div[{inside_div}]')
                                if ele.text != '':
                                    location = location + ele.text.replace("\n",'') + ","
                            except:
                                pass
                        at_this_location[i] = location
                        
                except: 
                    pass 
            
    except:
        pass
        
    
        
            
    
    #clear the search text field
    driver.find_element("id", "searchboxinput").clear()
    
    
```

```python
df['address'] = addresses
df['lat'] = latitude
df['long'] = longitude
df['at_this_location'] = at_this_location
```
