## Danish company data


```python
#imports
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', 500)
import math
#import for requests
import requests

#import beautiful soup
from bs4 import BeautifulSoup

#we will use selenium webdriver
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
import os
```

```python
#create webdriver instance
driver = webdriver.Chrome()

#get the url with login page
driver.get("https://www.proff.dk/segmentering?ef=2&et=52487&c=Aktieselskab&c=Anpartsselskab&c=Iv%C3%A6rks%C3%A6tterselskab&c=Forening%20eller%20selskab%20med%20begr%C3%A6nset%20ansvar&c=Kommanditaktieselskab%2FPartnerselskab&c=Kommanditselskab&c=Medarbejderinvesteringsselskab&c=Personligt%20ejet%20Mindre%20Virksomhed&c=Selskab%20med%20begr%C3%A6nset%20ansvar&samplerFilter=true")

sleep(1)
try:
    driver.find_element("xpath","/html/body/div[1]/div/div/div/div[2]/div/button[3]").click()
except:
    pass

links = []
soup = BeautifulSoup(driver.page_source, "html5")

for link in soup.find_all("div", class_="listing"):
    links.append(link.a['href'])
    
    
```

```python
links
```

```python
links[0] = links[0].replace("firma", "regnskab")
url = requests.get("https://www.proff.dk"+links[0])
soup = BeautifulSoup(url.content, "html5")
table = soup.find_all('table')
df = pd.read_html(str(table))
```

```python
df_final = pd.concat([pd.read_html(str(table))[1].transpose(), pd.read_html(str(table))[2].transpose()], axis=1)
```

```python
df_final
```

```python
#set index to the first columns and transpose
df1 = pd.read_html(str(table))[1].set_index("RESULTATREGNSKAB  i hele 1000").transpose()
```

```python
#set index to the first columns and transpose
df2 = pd.read_html(str(table))[2].set_index("BALANCEREGNSKAB  i hele 1000").transpose()
```

```python
#merge the two dataframes columnwise
df = pd.concat([df1,df2], axis=1)
```

```python
#print the datframe
df
```

```python
#export to excel

cwd = os.getcwd()

df.to_excel(cwd + "/DanishCompanydata.xlsx")
```
